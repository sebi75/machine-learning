Lab: Model Representation

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

In this specific laboratory we will use a small dataset of only 2 training examples.
We'll have the size of the house (x) and the price of the house (y).

```python
# x_train is the input variable (size in 1000s square feet)
# y_train is the target (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

We'll use "m" to denote the number of training examples, in this case m = 2.

```python
m = len(x_train) # number of training examples
print(f"Number of training examples: {m}")
```

Plotting data. We will use a scatter plot to visualize the data, since it has only two properties to plot (size of the house and price).

```python
# plot datapoints
plt.scatter(x_train, y_train, marker="x", c="r")
# set title
plt.title("Housing prices")
# set y axis label
plt.ylabel("Price in 1000s of dollars")
# set x axis label
plt.xlabel("Size in 1000s of square feet")
plt.show()
```

Let's create a function that computes the model output for given w, b and input x.

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    # creates an array of zeros with the same shape as x
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

```python
w = 100
b = 100
tmp_f_wb = compute_model_output(x_train, w, b)

# plot model prediction
plt.plot(x_train, tmp_f_wb, c="r", label="Prediction")
plt.scatter(x_train, y_train, marker="x", c="b", label="Actual Values")

plt.title("Housing prices")
plt.ylabel("Price in 1000s of dollars")
plt.xlabel("Size in 1000s of square feet")
plt.legend()
plt.show()
```

As seen, for values w = 100 and b = 100, the prediction line is not accurate.
By modifying and experimenting with different values for w and b, we can get a better prediction line.

```python
w = 200 # increase weight 2x from previous
b = 100
tmp_f_wb = compute_model_output(x_train, w, b)

# plot model prediction
plt.plot(x_train, tmp_f_wb, c="r", label="Prediction")
plt.scatter(x_train, y_train, marker="x", c="b", label="Actual Values")

plt.title("Housing prices")
plt.ylabel("Price in 1000s of dollars")
plt.xlabel("Size in 1000s of square feet")
plt.legend()
plt.show()
```

```python
# now let's predict with the weights that we found the price for a house with 1200 square feet
w = 200
b = 100
x_i = 1.2
house_cost = (w * x_i + b) * 1000
print(f"${house_cost:.0f} dollars")
```
